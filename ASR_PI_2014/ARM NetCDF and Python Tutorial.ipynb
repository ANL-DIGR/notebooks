# ARM NetCDF Files:

ARM stores its data in Unidata's netCDF file format:  http://www.unidata.ucar.edu/software/netcdf/
As described by Unidata, netCDF is a set of software libraries and self-describing, machine-independent data formats that support the creation, access, and sharing of array-oriented scientific data. In a nutshell, netCDF is a compact, self-describing, machine-independent binary (no big endian versus little endian) file format that is very useful for storing, accessing, and sharing scientific datasets.  <br>

In addition to the original Fortran and C libraries, many languages provide interfaces to NetCDF. Popular data analysis packages used around ARM and ASR are Matlab and interactive data language (IDL). These are both commercial packages, but a free (or inexpensive for some distributions) high-level language seeing growing use in ARM is Python.

## File name format:

The file names for ARM netCDF files are structured in a very specific format, which includes information about the site, instrument or VAP, facility, data level, start date, and start time in UTC.  

The files are named according to the following naming convention. All characters are lowercase except for facility indicator. Only “a-z”, “A-Z”, “0-9”, and “.” characters are allowed.
(sss)(inst)(qualifier)(temporal)(Fn).(dl).(yyyymmdd).(hhmmss).(cdf or nc) where:<br><br>
(sss) is the three letter ARM site identifier (e.g., sgp, twp, nsa, pgh, nim, ena, mag). The identifier is defined by a geographic reference or the IATA 3-letter airport code to indicate approximate location. Fixed sites are named after a geographic reference, while AMF deployments use the IATA code. Exceptions may be made for moving deployments such as ship and aircraft, or for large geographic areas for satellite data.
<br>

(inst) is the ARM instrument abbreviation (e.g., mwr, met, ecor, mpl), or the name of a ARM value-added product (VAP). The abbreviation is typically an acronym describing the instrument suite or VAP, and may describe the method for retrieving the measured or derived quantity. The instrument abbreviation must not end a with a number since this can be confused with the data temporal resolution descriptor or other optional descriptors following the instrument abbreviation.
<br>

(qualifier) is an optional qualifier that distinguishes these data from other data sets produced by the same instrument or VAP (e.g. avg, 1long). The optional qualifier may have one or more additional qualifiers describing a specific algorithm method or instrument specifics. This qualifier is used to describe monthly, yearly or annual files.
<br>

(temporal) is an optional description of data temporal resolution (e.g., 30m, 1h, 5s, 200ms, 14d). All temporal resolution descriptors require a units identifier. Accepted abbreviations include: ns=nanosecond, us=microsecond, ms=millisecond, s=second, m=minute, h=hour, d=day, mo=month, yr=year. It is recommended that the primary datastream for use by the end user not have a data integration period in the name. Time integration periods are converted to the lowest unit description. When possible default to minutes. Example: 60 seconds is labeled as “1m”, 60 minutes is labeled as “1h”.
<br>

(Fn) is the ARM facility designation. A facility is designated with a capital letter followed by one or two numbers not padded with zeros (e.g., S1, C1, E13, B4, M1, I4). Extended facilities around a central facility at the remote sites (excluding SGP) indicate the correlation to the central facility by matching the first of the two required numerical characters. Example: central facility TWP-C2 is related to extended facilities E20, E21, E22, while TWP-C1 is related to E10, E11, E12. External data-products that cover a large locale use the facility designation of X1, while data-products that are specific to an ARM facility follow the extended facility two numeral character naming. For example, external data associated specifically with the TWP-C1 facility would be named TWP-X11, TWP-X12, etc. Character coding designations: B = boundary, C = central, E = extended, I = intermediate, M = mobile, S = supplemental, X = external data site.

The file used in the first example below, sgpmetE13.b1.20140218.000000.cdf, includes the following information in the filename: <br>
site = sgp<br>
instrument = met<br>
qualifier = none provided<br>
facility ID = E13<br>
data level = b1<br>
start date = 20140218<br>
start time = 000000 UTC

## Typical File Contents

All ARM-standard netCDF files include some basic metadata that describe their contents. Each file contains one or more dimensions used to describe each variable by time, space, and or category. The majority of variables will be dimensioned by time, and time will always be the first dimension listed for multi-dimensional fields such as those commonly described by something such as time, height.<br>

The next section in each file lists the variables. Each variable has a name and data type such as int, double, or float. These variables are all further described by attributes. Typical attributes for each variable are the long_name, units, and missing_value, but there may be others such as valid_min, valid_max, and valid_delta often associated with automated quality control variables. 

Finally, the last section of each file contains the global_attributes. This section contains additional information that applies to the file as a whole instead of a single variable. This section typically contains information such as the software version used to create the file, a history attribute showing when the file was created, instrument serial numbers, etc. The global attributes may also include information about QC fields if global QC tests are applied to the data and/or references to peer-reviewed journal articles or ARM Tech Reports describing the algorithm used to produce the data.  It is always useful to scan through the global attributes at least once before using the data to gain a better understanding of the dataset.

## Creating a basic 2D profile plot with matplotlib

This example loads data from the sgpvceil25kC1.b1 datastream to plot backscatter from the Vaisala ceilometer. You will need to download the data file from <a href = 'http://dev.arm.gov/~monroe/ARM_netCDF_tutorial_files/sgpvceil25kC1.b1.20140211.000009.cdf'>HERE</a> and change value in the file_name variable below to point to the location in which you downloaded the file. 

```python
# import 3rd-party module to read netCDF files
from netCDF4 import Dataset

# import masked-array log10 function for backscatter
from numpy.ma import log10

# import matplotlib subplots function
from matplotlib.pyplot import subplots, colorbar

# Path to netCDF file, change /data_path/ to match where the file is located on your system
file_name = '/data_path/sgpvceil25kC1.b1.20140211.000009.cdf'

# use the Dataset function to create an object reference to the open netCDF file
file_obj2 = Dataset(file_name)

# store backscatter data into variable after taking the log of each value
backscatter = log10(file_obj2.variables['backscatter'][:])

# use the subplots function to create a single plot 
# this makes things easier as you move to multipanel plotting
fig, ax = subplots()

# print the shape of the backscatter array
print(backscatter.shape)

# you must transpose the array for standard 2D ARM data if you want time on the x-axis for plotting
backscatter = backscatter.transpose()

# use the pcolormesh method of the axis object to create the 2D plot, create a reference called pplot
pplot = ax.pcolormesh(backscatter)

# add a colorbar with the default color table - note that this works fine with a single plot
# you should use a different approach for multipanel plots, an example is provided in Advanced section after the tutorial
colorbar(pplot)

# optional commands for cleaning up the plot a bit:

# add gridlines
ax.grid()

# scale the x axis, there are 5400 time values in the file
ax.set_xlim(0,5399)

# scale the y axis, there are 252 range bins in the file 
ax.set_ylim(0,251)

# add a title
ax.set_title('Backscatter')
```

## Easily getting time from an ARM netCDF file into the proper format for plotting:

This example will use the file_obj reference to the sgpmetE13.b1 file in the basic section of this tutorial above.

```python
# import netCDF4 num2date function
from netCDF4 import num2date

# get an object reference to the time variable
time_obj = file_obj.variables['time']

# get the units from the time_obj
time_units = time_obj.getncattr('units')

# get the time data values
times = time_obj[:]

# convert the times into the native python datetime format using the num2date function
times = num2date(times,time_units)

print(times)
```

There is a handy built-in function called plot_date that works nicely with the time array we just constructed. Unfortunately, the pcolormesh function does not like the native python datetime format when passed as an agrument for the x-axis. We will need to further convert this to the native matplotlib date format. Fortunately, this is very easy and the matplotlib format will work with both plot_date and pcolormesh:

```python
# import matplotlib function to convert times to native matplotlib format
from matplotlib.dates import date2num

# convert the times
new_times = date2num(times)

# print the new times to the screen to show that they have been converted
print(new_times)
```

## Automatically wrapping text for plot titles and labels:

If you ever find your plot labels becoming too long, there is a nice way to automate splitting these strings with a newline based on a specified length. The python wrap function will take care of this for you, and will automatically split the string along word boundaries. Here is an example: 

```python
# import wrap function from standard libraries
from textwrap import wrap

label_str ='I am a really, really, really, long attribute name that should never have been put in a netCDF file'

# split the string at 55 characters, and join the components back together with a newline (\n)
y_label = '\n'.join((wrap(label_str, 55))) 
print(y_label)
```

## Making a two-panel figure with different plot types in each axis:

This example includes everything you need to make time, height profile plot from a ARM NetCDF file along with a secondary time series plot of a 1D variable in the second panel.
```python
# import wrap function from standard libraries
from textwrap import wrap

# import 3rd-party module to read netCDF files, and num2date function for converting time
from netCDF4 import Dataset, num2date

# import masked-array log10 function for backscatter
from numpy.ma import log10

# import matplotlib subplots and colorbar functions
from matplotlib.pyplot import subplots, colorbar,  cm

# import make_axes_locatable function to bettter position colorbar 
from mpl_toolkits.axes_grid1 import make_axes_locatable

# import matplotlib function to convert times to native matplotlib format, also date and hour formatters for plot labeling
from matplotlib.dates import date2num, DateFormatter, HourLocator

# Path to netCDF file, change /data_path/ to match where the file is located on your system
file_name = '/data_path/sgpvceil25kC1.b1.20140211.000009.cdf'

# use the Dataset function to create an object reference to the open netCDF file
file_obj2 = Dataset(file_name)

# Set date formatting for later use on the x-axis of each plot
hoursFmt= DateFormatter('%H')    
hours = HourLocator(list(range(0,24,3))) # every 3 hours   

# create a figure and axes objects for the plots
fig, (ax1,ax2) = subplots(1,2)
fig.set_size_inches(10,5)

# create an object reference to the backscatter and first cloud base height variables
backscatter_obj = file_obj2.variables['backscatter']
first_cbh_obj = file_obj2.variables['first_cbh']

# create an object reference to the range coordinate variable
range_obj = file_obj2.variables['range']

# create an object reference to the time variable
time_obj = file_obj2.variables['time']

# get data for plotting
backscatter = log10(backscatter_obj[:])
first_cbh_data = first_cbh_obj[:]
range_data = range_obj[:]
time_data = time_obj[:]
time_units = time_obj.getncattr('units')
# convert the times into the native python datetime format using the num2date function
times = num2date(time_data,time_units)
# convert the times to matplotlib format for plotting with pcolormesh function
times = date2num(times)


# get variable attributes for plot labels
backscatter_name = backscatter_obj.getncattr('long_name')
first_cbh_name = first_cbh_obj.getncattr('long_name')
first_cbh_units = first_cbh_obj.getncattr('units')
range_name = range_obj.getncattr('long_name')
range_units = range_obj.getncattr('units')

# start the first plot of backscatter ----------------------------------

# change from default colormap of "jet" for pcolormesh plot
color_map = cm.spectral

# create the pcolormesh plot on the first axis, but store a reference to the plot in pplot
pplot = ax1.pcolormesh(times,range_data,backscatter.transpose(), vmin=0, vmax=4, cmap=color_map)

# since we are using matplotlib-native dates, format the x-axis as dates
ax1.xaxis_date()

# scale the y-axis based on the min and max range data values
ax1.set_ylim(range_data.min(), range_data.max())

# create string for y-axis label, and use wrap function to wrap text to next line
# if over specified character length
ylabel = ''.join((range_name,' (',range_units,')'))
ylabel = '\n'.join((wrap(ylabel, 40))) 

# add a y-axis label to the plot
ax1.set_ylabel(ylabel, fontsize=10,fontweight='bold')

# add a x-axis label to the plot
ax1.set_xlabel('Time (GMT)', fontsize=10,fontweight='bold')

# add a title to the plot 
ax1.set_title(backscatter_name,fontsize=12,fontweight='bold')

# set limits for x axis 
ax1.set_xlim(times[0], times[-1])

# add a small area to the right of the main axes for the colorbar
# it is good to use this method if you are creating a multipanel plot
# and using the tight_layout function for plot spacing
divider = make_axes_locatable(ax1)
cax = divider.append_axes("right", "3%", pad="3%")

# create the colorbar
# references the pcolormesh plot and previously appended axes area
cb = colorbar(pplot,cax=cax)

# change font size and weight of colorbar labels
for label in cb.ax.get_yticklabels():
    label.set_fontsize(8)
    label.set_fontweight('bold')

# add label to colorbar
cb.ax.set_ylabel('Amplitude (log of counts)',fontsize=8,fontweight='bold')

# add gridlines to the plot
ax1.grid()

# set the size and weight of yticks on the plot
for tick in ax1.yaxis.get_major_ticks():
    tick.label.set_fontsize(8)
    tick.label.set_fontweight('bold')

# set the size and weight of xticks on the plot
for tick in ax1.xaxis.get_major_ticks():
    tick.label.set_fontsize(8)
    tick.label.set_fontweight('bold')

# manually format the dates on the x axis
ax1.xaxis.set_major_locator(hours)
ax1.xaxis.set_major_formatter(hoursFmt)

# start the second plot of first cloud base height ----------------------------------

# create the pcolormesh plot on the first axis, but store a reference to the plot in pplot
ax2.plot_date(times,first_cbh_data,'ko',mec='k',ms=3)

# create string for y-axis label, and use wrap function to wrap text to next line
# if over specified character length
ylabel = ''.join(('Height (',first_cbh_units,')'))
ylabel = '\n'.join((wrap(ylabel, 40))) 

# add a y-axis label to the plot
ax2.set_ylabel(ylabel, fontsize=10,fontweight='bold')

# add a x-axis label to the plot
ax2.set_xlabel('Time (GMT)', fontsize=10,fontweight='bold')

# add a title to the plot 
ax2.set_title(first_cbh_name,fontsize=12,fontweight='bold')

# set limits for x axis
ax2.set_xlim(times[0], times[-1])

# set limits for y axis to match backscatter
ax2.set_ylim(0, 7545.)

# add gridlines to the plot
ax2.grid()

# set the size and weight of yticks on the plot
for tick in ax2.yaxis.get_major_ticks():
    tick.label.set_fontsize(8)
    tick.label.set_fontweight('bold')

# set the size and weight of xticks on the plot
for tick in ax2.xaxis.get_major_ticks():
    tick.label.set_fontsize(8)
    tick.label.set_fontweight('bold')

# manually format the dates on the x axis
ax2.xaxis.set_major_locator(hours)
ax2.xaxis.set_major_formatter(hoursFmt)
        
# create as super title for both plots
fig.suptitle('sgpvceil25kC1.b1 for 20140211',fontsize=14,fontweight='bold') 

# use tight layout function to format white space between axes, etc.
fig.tight_layout(pad=3,h_pad=4)
```

## Working with masked arrays:

The numpy masked array module provides a nice mechanism for masking out undesirable values in a data array.  This is very useful for masking out ARM missing values (typically -9999), NaNs (not a number), etc. By default, the python netCDF4 module typically checks for the existence of a missing_value or __FillValue variable attribute and masks these values from the array when data are read from netCDF file.  This means that data loaded from a netCDF file is usually a numpy masked array.  You may have to use the type command to check for this on your system.  If the mask is not added automatically you must do it yourself.<br><br>

A masked array technically stores two arrays in memory.  One array is the actual data, and the other array is a mask with values of 0 or 1. Values of 0 (false) indicate that the value in the corresponding data array is not masked, while values of 1 (true) indicate that the value in the data array should be masked from display, computations, and plotting. 

```python
# import numpy masked array module
from numpy import ma

masked_array = ma.array([1,2,-9999,4,-9999,6,99999],mask=[0,0,1,0,1,0,1])

# print array showing masked values
print masked_array
```

## Decoding Bit-packed ARM QC Tests and Adding Flagged Values to Data Mask:

Many ARM QC variables store the results in a binary bit-packed format that allows as single value to represent multiple states at the same time. Here is a basic example of the standard missing_value, valid_min, valid_max, and valid_delta QC tests applied to many ARM instrument variables:<br><br>

qc_bit_1_description = "Value is equal to missing_value." ;<br>
qc_bit_1_assessment = "Bad" ;<br>
qc_bit_2_description = "Value is less than the valid_min." ;<br>
qc_bit_2_assessment = "Bad" ;<br>
qc_bit_3_description = "Value is greater than the valid_max." ;<br>
qc_bit_3_assessment = "Bad" ;<br>
qc_bit_4_description = "Difference between current and previous values exceeds valid_delta
<br>
<br>

If you simply want to know if the data passed all automated QC tests, you can simply check the QC version of the variable for times when the values equal 0. However, if you will need to unpack the individual bit numbers to check if a specific test failed.<br><br>
For this example, you will need to download the enaskyrad60sC1.b1 data file from <a href= 'http://dev.arm.gov/~monroe/ARM_netCDF_tutorial_files/enaskyrad60sC1.b1.20140212.000000.cdf'>HERE</a>.  We will open this file, grab data from both the sky_ir_temp and qc_ir_temp variables, and set the mask to filter out the data points that failed the valid_min QC test described in bit_2 above.  I will manually set the mask in this example to filter out values equal to -9999 if they exist:


```python
# import 3rd-party module to read netCDF files
from netCDF4 import Dataset

# import numpy where function to search array, and ma module for masked arrays
from numpy import where, ma

# import zeros function to initialize mask if needed
from numpy import zeros

# Path to netCDF file, change /data_path/ to match where the file is located on your system
file_name = '/data_path/enaskyrad60sC1.b1.20140212.000000.cdf'

# use the Dataset function to create an object reference to the open netCDF file
file_obj3 = Dataset(file_name)

# get reference to sky_ir_temp variable object so we can get more information later
sky_ir_temp_obj = file_obj3.variables['sky_ir_temp']

# get sky_ir_temp data and qc_sky_ir_temp data
sky_ir_temp = sky_ir_temp_obj[:]
qc_sky_ir_temp = file_obj3.variables['qc_sky_ir_temp'][:]

# here is an easy way to check if the array is already masked and convert if needed
#  this is only necessary if you don't already have a masked array with a full mask
if not hasattr(sky_ir_temp,'mask'):
    sky_ir_temp = ma.masked_equal(sky_ir_temp,-9999) # mask -9999 values by default

    # if no -9999 values, mask will just be a value of False, not an array
    if sky_ir_temp.mask.dtype.name.startswith('bool'):
        sky_ir_temp.mask = zeros(sky_ir_temp.shape)
    
# get the times when the valid_min test (bit 2) failed in qc_sky_ir_temp
bit_number = 2
bit_check = 2**(bit_number-1) & qc_sky_ir_temp # use bitwise operator to check entire array  

# any value > 0 means that bit 2 was tripped and the data failed the valid_min test
tripped = where(bit_check > 0)
if tripped[0].size > 0:
    print(' '.join(('Bit',str(bit_number),'was tripped!')))
    
# display the index values in the array that were tripped
print(tripped)

# add times that data failed the valid_min test to mask
sky_ir_temp.mask[tripped] = 1
```

Now that the mask is set, let's plot both the unmasked and masked data to see the difference (You will also see that the automated QC wasn't perfect):

```python
# import matplotlib subplots 
from matplotlib.pyplot import subplots

# create a plot with 2 subplots, 2 rows, 1 column
fig, (ax1, ax2) = subplots(2,1)
fig.set_size_inches(10,10)

# first plot all of the data without the mask
ax1.plot(sky_ir_temp.data)
ax1.set_title('Without QC')

# next plot the data with the bad values masked out
ax2.plot(sky_ir_temp,'r')
ax2.set_title('With QC')
```

## Generating a range of dates to loop over:

Generating a desired range of dates can be very useful for many applications, such as reading/writing multiple files or creating multiple daily plots. This example will demonstrate how easy it is to accomplish this task using the 3rd-party python dateutil module (https://pypi.python.org/pypi/python-dateutil), which is included with matplotlib or can be installed separately. Example code is below:

```python
from dateutil.parser import parse
from dateutil.rrule import rrule, DAILY

# set a start date and end date (can use string or integer since str() function used below)
startdate = '20131215'
enddate = '20140115'

print(parse(startdate))

# generate list of dates in python datetime format
alldates = list(rrule(DAILY,dtstart=parse(str(startdate)),until=parse(str(enddate))))

# print the dates 
print(alldates)

# convert the list of datetime dates into a list of date strings
strdates = [date.strftime('%Y%m%d') for date in alldates]

# print string dates
print(strdates)
```

## Creating a multipanel plot by looping over variable names:

```python
# must insert this statement to render the plots within the notebook
# this is specific to the ipython notebook
%matplotlib inline

# import 3rd-party module to read netCDF files
from netCDF4 import Dataset, num2date
from matplotlib.pyplot import subplots
from matplotlib.dates import DateFormatter, HourLocator

# Path to netCDF file, change /data_path/ to match where the file is located on your system
file_name = '/data_path/sgpmetE13.b1.20140218.000000.cdf'

# use the Dataset function to create an object reference to the open netCDF file
file_obj = Dataset(file_name)

# get the time data into the times array for later plotting
time_obj = file_obj.variables['time']
time_units = time_obj.getncattr('units')
times = time_obj[:]
times = num2date(times,time_units)

# create a list of variable names to loop over
variables = ['temp_mean','rh_mean','atmos_pressure']

# initialize some plotting parameters for each variable
plot_color = {}
plot_color['temp_mean'] = 'r'
plot_color['rh_mean'] = 'b'
plot_color['atmos_pressure'] = 'g'

plot_symbol = {}
plot_symbol['temp_mean'] = 'o'
plot_symbol['rh_mean'] = '+'
plot_symbol['atmos_pressure'] = '|'

# initialize your axes reference as a dictionary
ax = {}

# use variable names as the keys in your axes dictionary
# sharex=True means shared x axes, only bottom plot has xticks
fig, (ax['temp_mean'], ax['rh_mean'], ax['atmos_pressure']) = subplots(3,1,sharex=True)
fig.set_size_inches(10,10)

# Set date formatting for later
hoursFmt= DateFormatter('%H')    
hours = HourLocator(list(range(0,24,3))) # every 3 hours    

# loop over variable names and begin plotting
for var in variables:
    
    # get variable data and attribute with each pass
    var_obj = file_obj.variables[var]
    var_data = var_obj[:]
    var_long_name = var_obj.getncattr('long_name')
    var_units = var_obj.getncattr('units')
    
    # plot each variable on its axis
    ax[var].plot_date(times, var_data, ''.join((plot_color[var],plot_symbol[var])),
        mec=plot_color[var],ms=4.0)
    
    # set ylabel and title for each axis
    ax[var].set_ylabel(''.join((var,'(',var_units,')')), fontweight='bold')
    ax[var].set_title(var_long_name, fontweight='bold')
    
    # add a grid to each axis
    ax[var].grid()
    
    # set the size and weight of yticks on each axis (plot)
    for tick in ax[var].yaxis.get_major_ticks():
        tick.label.set_fontsize(8)
        tick.label.set_fontweight('bold')
    
    '''We are setting this for the last variable or bottom axis since it is the only 
    one with time labels due to the shared xaxis set above'''    
    if var == variables[-1]:
        
        # manually format the date axis
        ax[var].xaxis.set_major_locator(hours)
        ax[var].xaxis.set_major_formatter(hoursFmt)
            
        # set the size and weight of xticks on the plot
        for tick in ax[var].xaxis.get_major_ticks():
            tick.label.set_fontsize(8)
            tick.label.set_fontweight('bold')
            
        # add a x-axis label to the plot
        ax[var].set_xlabel('Time (GMT)', fontsize=10,fontweight='bold')
    
```
