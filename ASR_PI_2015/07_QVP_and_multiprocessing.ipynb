Lets define our function to be mapped to the cluster from our previous tutorial! (and do relevant imports)

```python
# turn on plotting
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from netCDF4 import num2date
import matplotlib.dates as mdates
#utilities for interacting with the OS
import os

import pyart
def retrieve_qvp(filename, flds = None):
    if flds == None:
        flds = ['differential_phase', 
            'cross_correlation_ratio',
            'spectrum_width',
            'reflectivity', 'differential_reflectivity']
    radar = pyart.io.read(filename)
    desired_angle = 20.0
    index = abs(radar.fixed_angle['data'] - desired_angle).argmin()
    qvp = {}
    for fld in flds:
        this_fld = radar.get_field(index, fld)[:, 0:900].mean(axis = 0)
        qvp.update({fld:this_fld})
    qvp.update({'range': radar.range['data'][0:900], 'time': radar.time})
    x,y,z = pyart.io.common.radar_coords_to_cart(qvp['range']/1000.0, 0.0, 
                                             radar.fixed_angle['data'][index])
    qvp.update({'height': z})
    del radar
    return qvp
```

```python
indir = '/data-in/radar/sgp/sgpcsaprsur/sur/20110520/'
all_csapr_files = os.listdir(indir)
all_csapr_files.sort()
interesting_files = []
for fn in all_csapr_files[20:180]:
    if 'mdv in fn':
        interesting_files.append(indir + fn)
```

```python
#print interesting_files
print len(interesting_files)
```

Here is where the Magic happens! We will use the idea of Map-Reduce to map the problem to our cluster. 
http://en.wikipedia.org/wiki/MapReduce

First step imports Py-ART onto all the worker kernels. We then map the code and the inputs to the cluster.. We then reduce the problem via the result.get()


```python
#Turn off blocking so all engines can work async
My_View.block = False

#on all engines do an import of Py-ART
My_View.execute('import pyart')

#Map the code and input to all workers
result = My_View.map_async(retrieve_qvp,interesting_files)

#Reduce the result to get a list of output
qvps = result.get()
```

This will return a list of QVP dictionaries

Lets plot a random one!

```python
plt.plot(qvps[10]['reflectivity'], qvps[10]['height']/1000.0)
plt.xlabel('mean reflectivity(dbZ)')
plt.ylabel('Height above ground (km)')
plt.ylim([0,20])
```

Super cool! Looks like an anvil was over us at the time!

Now we are going to make a simple snippet that will take all the QVP single profiles and construct a time series. 

```python
#create an empty dictionary to be filled
time_series = {}

#work out how many profiles
n_times = len(qvps)

#Loop over all fields 
for fld in ['differential_phase', 
            'cross_correlation_ratio',
            'spectrum_width',
            'reflectivity', 'differential_reflectivity']:
    #Create empty time series
    this_fld = np.zeros([n_times, len(qvps[0][fld])])
    
    #create time array
    times = np.zeros(n_times)
    
    #Loop over all times
    for i in range(n_times):
        
        #Create a datetime object
        dateobj = num2date(qvps[i]['time']['data'][0], qvps[i]['time']['units'])
        
        #Append a numerical date using Matplotlib's time reference
        times[i] = mdates.date2num(dateobj)
        
        #Update the profile to the time series
        this_fld[i,:] = qvps[i][fld]
    
    #append the time series to the dictionary
    time_series.update({fld: this_fld[times.argsort(), :]})
times.sort()
```

Lets plot! 

```python
fig = plt.figure(figsize = [15,5])
pc = plt.pcolormesh(times, qvps[0]['height']/1000.0, 
                    time_series['reflectivity'].transpose(),
                    vmin = -20, vmax = 64)
ax = plt.gca()
ax.xaxis_date()
#Magically deal with time!
fig.autofmt_xdate()
plt.xlabel('Time (UTC)')
plt.ylabel('Height (km)')
plt.ylim([0,20])
cb = plt.colorbar()
cb.set_label('Mean Reflectivity factor (dBZ)')
```

```python
times.sort()
fig = plt.figure(figsize = [15,5])
pc = plt.pcolormesh(times, qvps[0]['height']/1000.0, 
                    time_series['differential_reflectivity'].transpose(),
                    vmin = -3, vmax = 7)
ax = plt.gca()
ax.xaxis_date()
fig.autofmt_xdate()
plt.xlabel('Time (UTC)')
plt.ylabel('Height (km)')
plt.ylim([0,20])
cb = plt.colorbar()
cb.set_label('Mean Differential Reflectivity (dB)')
```

```python
times.sort()
fig = plt.figure(figsize = [15,5])
pc = plt.pcolormesh(times, qvps[0]['height']/1000.0, 
                    time_series['spectrum_width'].transpose(), vmax = 6)
ax = plt.gca()
ax.xaxis_date()
fig.autofmt_xdate()
plt.xlabel('Time (UTC)')
plt.ylabel('Height (km)')
plt.ylim([0,20])
cb = plt.colorbar()
cb.set_label('Mean Spectral Width (ms$^{-1}$)')
```

```python
times.sort()
fig = plt.figure(figsize = [15,5])
pc = plt.pcolormesh(times, qvps[0]['height']/1000.0, 
                    time_series['cross_correlation_ratio'].transpose())
ax = plt.gca()
ax.xaxis_date()
fig.autofmt_xdate()
plt.xlabel('Time (UTC)')
plt.ylabel('Height (km)')
plt.ylim([0,20])
cb = plt.colorbar()
cb.set_label('Mean Correlation coeff')
```
