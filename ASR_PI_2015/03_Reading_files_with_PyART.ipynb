# Reading radar files with Py-ART and the radar object

In this notebook we will examine how to read radar files in Py-ART and how Py-ART stores radar volumes.  For this we will be examining a file from the W-band Scanning ARM Cloud Radar ([WSACR](http://www.arm.gov/instruments/wsacr)).  

Execute the cells below by clicking on them and hiting cltr+enter


```python
# turn on interactive plotting
%matplotlib inline
import matplotlib.pyplot as plt
```

First we need to import Py-ART, this allows us to use the routine to read in data from the netCDF file.

```python
import pyart
```

```python
radar = pyart.io.read('data/sgpwsacrcwrhiC1.a1.20120820.204016.nc')
```

We have just created a **Radar** object which is the class Py-ART uses to store and process radar volume data.  Use Python's built in help system to examine this object:

```python
type(radar)
```

```python
help(radar)
```

We can also use IPython's help system.

```python
radar?
```

---
This tells us about the Radar class in general.  We can use the `info` class method to describe the data in the file we have read. 

```python
radar.info('compact')
# try out 'standard' or 'full' also 
```

---
For convience the *radar_info* command can provide this data from a command line.

```python
# ! indicate that a command will be execute from a shell not IPython
# The results are printed here. 
!radar_info --compact data/sgpwsacrcwrhiC1.a1.20120820.204016.nc
```

---
Notice from the above infomation we are looking at a Range Height Indicator or RHI scan.  This means for each sweep in the volume the azimuth angle is fixed and the radar samples a number of elevation angles.  The information on these elevation angles is stored in the **elevation** attribute as a Python dictionary.

```python
type(radar.elevation)
```

```python
radar.elevation.keys()
```

```python
print radar.elevation['long_name']
print radar.elevation['standard_name']
print radar.elevation['units']
```

```python
radar.elevation['data']
```

We can easily plot the elevation angles for all the sweeps in the volume.

```python
plt.plot(radar.elevation['data'])
```

From the local extremal in the graph above this volume appears to contain 17 sweeps.  We can check to see that this agrees with the number of sweep reported by the file:

```python
radar.nsweeps
```

---
Sweeps are typically the unit of radar data that we want to work with.  The indices which delinate a particular sweeps are stores in the **sweep_start_ray_index** and **sweep_end_ray_index** attribute of the radar object.  Using these to access sweep data can be cumbersome and care must be taken to avoid off-by-one errors.  For convience the radar object provide a get_slice method which will return slice object which can be used to select out the rays belonging to a particular sweep.

```python
radar.get_slice?
```

---
Using this functionality to plot the elevation angles from the first two sweeps in the volume.

```python
sweep_0_slice = radar.get_slice(0)
plt.plot(radar.elevation['data'][sweep_0_slice])
plt.draw()
```

```python
sweep_1_slice = radar.get_slice(1)
plt.plot(radar.elevation['data'][sweep_1_slice])
plt.draw()
```

---
The radar fields or moments are stored in the **fields** attribute as a dictionary of dictionaries.  

```python
radar.fields.keys()
```

```python
type(radar.fields['reflectivity'])
```

```python
radar.fields['reflectivity'].keys()
```

```python
print radar.fields['reflectivity']['standard_name']
print radar.fields['reflectivity']['units']
print radar.fields['reflectivity']['coordinates']
```

---
We can plot this data as in a b-scan representation.

```python
plt.figure(figsize=(20, 10))
plt.imshow(radar.fields['reflectivity']['data'], aspect=0.1, origin='bottom')
plt.xlabel('range gate')
plt.ylabel('ray number')
```

---
Using the sweep slice object to look at only the second sweep.

```python
plt.figure(figsize=(10, 10))
plt.imshow(radar.fields['reflectivity']['data'][sweep_1_slice], aspect=0.5, origin='bottom')
plt.xlabel('range gate')
plt.ylabel('ray number')
```
