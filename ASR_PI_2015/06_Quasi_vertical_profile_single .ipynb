# Extracting a Quasi-Vertical Profile from CSAPR using Py-ART

In this notebook we will use Py-ART to look at polarimetric data from a tilt of a C-SAPR PPI

Execute the cells below by clicking on them and hiting cltr+enter

```python
# turn on inline plotting
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#import Py-ART
import pyart
```

```python
#we will be using a data file from the ARM data cluster.. 
#Replace the listing below to point to a file on your computer
in_dir = '/data-in/radar/sgp/sgpcsaprsur/sur/20110520/'
in_file = '110635.mdv'
filename = in_dir + in_file
```

Read the C-SAPR data into our radar object

```python
radar = pyart.io.read(filename)
```

Examine the file contents

```python
radar.fields.keys()
```

Lets jump straight to viewing a PPI

```python
my_display = pyart.graph.RadarDisplay(radar)
my_display.plot_ppi('reflectivity', sweep = 1, vmin = -8, vmax = 64)
```

now we need to pick the tilt that is at the desired elevation angle

```python
angle_wanted = 20.0
index = abs(radar.fixed_angle['data'] - angle_wanted).argmin()
print index, radar.fixed_angle['data'][index]
```

so now we want to isolate only the moments when the antenna is at the desired elevation

```python
start_index = radar.sweep_start_ray_index['data'][index]
end_index = radar.sweep_end_ray_index['data'][index]
print start_index, end_index, radar.fields['reflectivity']['data'].shape
fig = plt.figure(figsize = [15,6])
plt.pcolormesh(radar.fields['reflectivity']['data'][start_index:end_index, :].transpose())
```

```python
qvp = {}
for fld in ['differential_phase', 
            'cross_correlation_ratio',
            'normalized_coherent_power',
            'spectrum_width',
            'reflectivity', 'differential_reflectivity']:
    this_fld = radar.fields[fld]['data'][start_index:end_index, 0:900].mean(axis = 0)
    qvp.update({fld:this_fld})
qvp.update({'range': radar.range['data'][0:900], 'time': radar.time})
```

```python
plt.plot(qvp['reflectivity'], qvp['range']/1000.0)
plt.xlabel('mean reflectivity(dbZ)')
plt.ylabel('Range from radar (km)')
```

Range from the radar is somewhat meaningless, lets use 4/3 $R_e$ to do height above ground

```python
x,y,z = pyart.io.common.radar_coords_to_cart(qvp['range']/1000.0, 0.0, 
                                             radar.fixed_angle['data'][index])
qvp.update({'height': z})
```

```python
plt.plot(qvp['reflectivity'], qvp['height']/1000.0)
plt.xlabel('mean reflectivity(dbZ)')
plt.ylabel('Height above ground (km)')
plt.ylim([0,20])
```

```python
plt.plot(qvp['differential_reflectivity'], qvp['height']/1000.0)
plt.xlabel('mean Z$_{dr}$(db)')
plt.ylabel('Height above ground (km)')
plt.ylim([0,20])
```

Lets now put this in the form of a function

```python
def retrieve_qvp(radar, desired_angle):
    index = abs(radar.fixed_angle['data'] - desired_angle).argmin()
    qvp = {}
    for fld in ['differential_phase', 
            'cross_correlation_ratio',
            'normalized_coherent_power',
            'spectrum_width',
            'reflectivity', 'differential_reflectivity']:
        this_fld = radar.fields[fld]['data'][start_index:end_index, 0:900].mean(axis = 0)
        qvp.update({fld:this_fld})
    qvp.update({'range': radar.range['data'][0:900], 'time': radar.time})
    x,y,z = pyart.io.common.radar_coords_to_cart(qvp['range']/1000.0, 0.0, 
                                             radar.fixed_angle['data'][index])
    qvp.update({'height': z})
    return qvp
```

```python
radar2 = pyart.io.read(in_dir + '121634.mdv')
my_display = pyart.graph.RadarDisplay(radar2)
my_display.plot_ppi('reflectivity', sweep = 1, vmin = -8, vmax = 64)
```

```python
qvp2 = retrieve_qvp(radar2, 20.0)
```

```python
plt.plot(qvp2['reflectivity'], qvp2['height']/1000.0)
plt.xlabel('mean reflectivity(dbZ)')
plt.ylabel('Height above ground (km)')
plt.ylim([0,20])
```

```python
plt.plot(qvp2['differential_reflectivity'], qvp2['height']/1000.0)
plt.xlabel('mean Z$_{dr}$(db)')
plt.ylabel('Height above ground (km)')
plt.ylim([0,20])
```
